Series Index set


```python
marks=[67,87,56.0,87]
subject=['bangla','math','ict','eng']
c=pd.Series(marks,index=subject)
```

Series Name

```python
mark=[50,80,60,60,70]
subject=['bangla','english','Ict','math','science']
b=pd.Series(mark,index=subject,name='student mark list')
```

Series from dict


```python
mark={'math':76,'english':87,'science':60,'hindi':100}
a=pd.Series(mark,name='student mark list')
```
